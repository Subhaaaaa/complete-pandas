Pandas Series :
A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

```python
import numpy as np
import pandas as pd
```

```python
# String
countries = ['india','Srilanka','Pakistan','USA']
pd.Series(countries)
```

```python
# Custom index and naming the series
marks = [12,122,234,232,121]
subjects = ['Maths','Physics','Chemistry','Biology','Computer']
pd.Series(marks,index=subjects, name = 'Mark list')

```

```python
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}

marks_series = pd.Series(marks, name = 'Subhu ke marks')
marks_series
```

Important Series Attributes

size,dtype,name,is_unique,index,values
